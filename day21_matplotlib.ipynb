```python
import matplotlib.pyplot as plt
import numpy as np 
# to show chart in jupyter notebook
%matplotlib inline
```

```python
x = np.linspace(0,10,100) # generate 100 points from 0 to 10
y = np.sin(x)  # sine wave
```

```python
plt.plot(x,y,color ='black')
plt.title("This is Sine wave",fontsize = 14)
plt.ylabel("Sine values")
plt.xlabel('random numbers')
plt.show()
```

```python
# create a cos wave
y2 = np.cos(x)
```

```python

plt.plot(x,y2,color = 'green')
plt.title("This is cosine wave", fontsize = 16)
plt.xlabel("Random Numbers")
plt.ylabel("Cosine Values")
plt.show()
```

```python
plt.figure(figsize = (8,5))
plt.plot(x,y,color ='black',label = 'sine wave line')
plt.plot(x,y2,color = 'green',label = 'cos wave line')
plt.title("Wave form trending line", fontsize = 16)
plt.xlabel("Random Numbers")
plt.ylabel("Wave Values")
plt.legend()
plt.savefig('chart1.png')
plt.show()
```

Exercise 5: Sales Forecast Visualization (Shaded Area Plot)

Objective: Visualize historical sales along with forecasted values using an area plot.

Task: Create an area plot showing actual vs. forecasted sales. Customization: Use different colors for actual and forecasted values. Add hatching patterns for the forecasted area.

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
sales = [5000, 5200, 5400, 5300, 5500, 5700]
forecast = [5800, 5900, 6000, 6100, 6200, 6300]
```

```python
plt.fill_between(months,sales,label='Actual Sales',color='green',alpha = 0.5)
plt.fill_between(months,forecast,label='Forecasting Sales',color='blue',alpha = 0.4,hatch='/')
plt.title("Actual vs Forecasted Sales in $")
plt.legend(loc='upper right')
plt.grid(axis = 'y',linestyle='--',alpha =0.4)
plt.show()
```

Final Task: Correlation between Sales and Advertisement Spend

Objective: Explore the relationship between sales and advertising spend using a scatter plot.

Task: Create a scatter plot of sales vs. ad spend.
Customization:
Add a trendline.
Customize marker size and color based on ad spend.

```python
ad_spend = [1000, 1500, 1800, 2000, 2300, 2500, 3000, 3200]
sales = [4000, 4500, 5000, 5200, 5800, 6000, 6400, 7000]
```

```python
plt.figure(figsize=(8,6))
plt.scatter(ad_spend,sales,c=ad_spend,cmap='cool',
            s=[spend/10 for spend in ad_spend],
            alpha=0.7,edgecolors='black')

# to add a trendline
z = np.polyfit(ad_spend,sales,1)
p = np.poly1d(z)
plt.plot(ad_spend,p(ad_spend),linestyle = '--',color = 'red',label= 'trendline')

# customize the plot
plt.title('Sales vs Advertisment Spend',fontsize= 16)
plt.xlabel('Advertisment Spend')
plt.ylabel('Sales')
plt.legend()

plt.show()
```

```python
# Data: Advertisement Spend and Sales
ad_spend = [1000, 1500, 1800, 2000, 2300, 2500, 3000, 3200]
sales = [4000, 4500, 5000, 5200, 5800, 6000, 6400, 7000]

# Create the scatter plot
plt.figure(figsize=(8, 6))
sc = plt.scatter(ad_spend, sales, 
                 c=ad_spend, cmap='cool',  # Color based on ad spend
                 s=[spend / 10 for spend in ad_spend],  # Marker size based on ad spend
                 alpha=0.7, edgecolor='black')

# Add a trendline
z = np.polyfit(ad_spend, sales, 1)  # Linear regression
p = np.poly1d(z)
plt.plot(ad_spend, p(ad_spend), linestyle='--', color='red', label='Trendline')

# Customize the plot
plt.title('Sales vs Advertisement Spend', fontsize=16)
plt.xlabel('Advertisement Spend ($)', fontsize=12)
plt.ylabel('Sales ($)', fontsize=12)
plt.colorbar(sc, label='Ad Spend')  # Add color legend
plt.legend()

# Display the plot
# plt.savefig('test1.png')
plt.show()
```

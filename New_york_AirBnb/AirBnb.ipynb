### Data Cleaning

```python
df.isnull().sum()
```

```python
# dropping all missing values rows
df.dropna(inplace = True)
df.isnull().sum()
```

```python
# dealing with duplicated rows
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace = True)
df.duplicated().sum()
```

```python
df.info()
```

```python
df.dtypes
```

    from the above results , we can see there are no missing values
    and no duplicated rows but numeric
    we will convert id and host_id to object

```python
df['id'] = df['id'].astype(object)
df['host_id'] = df['host_id'].astype(object)
```

#### EDA

```python
# Univariate Analysis
df1 = df[df['price'] < 1500]
sns.boxplot(data =df1,x = 'price')
```

```python
# Price distribution
plt.figure(figsize=(8,5))
sns.histplot(data= df1,x = 'price',bins =100)
plt.title('Price Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
df.columns
```

```python
# Price distribution

plt.figure(figsize=(6,3))
sns.histplot(data = df1,x = 'availability_365',bins =100)
plt.title('availability_365 Distribution')
plt.ylabel('Frequency')
plt.show()
```

```python
price_group = df.groupby(df['neighbourhood_group'])['price'].mean()
```

```python
price_group
```

```python
# bar chart for average prices

plt.figure(figsize=(6,3))
sns.barplot(x = price_group.index,y = price_group.values)
plt.title('Average Price by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price')
plt.show()
```

```python
# price per bed
df['price_per_bed'] = df['price'] / df['beds']
df.head()
```

```python
# price per bed 

plt.figure(figsize=(6,3))
sns.barplot(x= df['neighbourhood_group'],y = df['price_per_bed'],errorbar=None)
plt.title('Average Price per Bed by Neighbourhood Group')
plt.xlabel('Neighbourhood Group')
plt.ylabel('Average Price per Bed')
plt.show()
```

```python
df.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

```python
sns.barplot(data = df,x = 'neighbourhood_group',y = 'price',hue='room_type',errorbar=None)
plt.show()
```

```python
# number of reviews and price rel
plt.figure(figsize=(8,5))
sns.scatterplot(data=df1, x = 'number_of_reviews',y = 'price',hue='neighbourhood_group')
plt.title('Locality and Reiview Dependency')
plt.show()
```
